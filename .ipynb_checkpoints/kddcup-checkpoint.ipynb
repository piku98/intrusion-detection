```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('./data/kddcup99_csv.csv')
```

```python
df.head()
```

```python
def create_unique_dict(colname):
    values = df[colname].unique()
    values_dict = {}
    for i in range(len(values)):
        values_dict[values[i]] = i
    return values_dict
```

```python
df['protocol_type'] = df['protocol_type'].replace(create_unique_dict('protocol_type'))
```

```python
df['service'] = df['service'].replace(create_unique_dict('service'))
```

```python
df['flag'] = df['flag'].replace(create_unique_dict('flag'))
```

```python
df['label'] = df['label'].replace(create_unique_dict('label'))
```

```python
df.head()
```

```python
df.T
```

```python
cols = df.columns
for i in range(len(cols)):
    if len(df[cols[i]].unique()) < 2:
        print(cols[i])
```

```python
df = df.drop(['lnum_outbound_cmds', 'is_host_login'], axis=1)
```

```python
corr = df.corr()
corr
```

```python
sb.set(rc={'figure.figsize':(16, 12)})
sb.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
removable_columns = []
for i in range(len(corr)):
    for j in range(len(corr)):
        if i != j and corr.iloc[i, j] > 0.99:
            print(corr.index[i], corr.columns[j])
            removable_columns.append(corr.columns[j])
```

```python
removable_columns
```

```python
df = df.drop(removable_columns, axis=1)
```

```python
df.head()
```

```python
X_data = df.drop(['label'], axis=1)
Y_data = df['label']
```

```python
scaler = StandardScaler()
X_data_scaled = scaler.fit_transform(X_data)
```

```python
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(X_data_scaled)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['p_component_' + str(i) for i in range(20)])
```

```python
principalDf.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(principalDf, Y_data, test_size=0.2, random_state=0, shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
```

```python
predictions = tree.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
model.fit(X_train, y_train)
```

```python
rf_predictions = model.predict(X_test)
rf_predictions
```

```python
accuracy_score(y_test, rf_predictions)
```

```python
principal_component_combined_shuffled_X = X_train 
principal_component_combined_shuffled_X = principal_component_combined_shuffled_X.append(X_test)
combined_shuffled_y = y_train
combined_shuffled_y = combined_shuffled_y.append(y_test)
```

```python
principal_component_combined_shuffled_df = principal_component_combined_shuffled_X.assign(label=combined_shuffled_y)
```

```python
principal_component_combined_shuffled_df.head()
```

```python
principal_component_combined_shuffled_df.to_csv('./data/principal_component_combined_shuffled_df')
```
